````python
import sys
sys.path.append('/usr/local/lib/python3.10/dist-packages/matlabengine-24.2-py3.10.egg')
import matlab.engine

# Start MATLAB engine and create a new Simulink model
eng = matlab.engine.start_matlab()
model = 'twoBusModel'
eng.new_system(model, nargout=0)               # Create new model
eng.open_system(model, nargout=0)

# Add components to the model:
# 1. Three-phase AC Source for Bus 1 (generator)
eng.add_block('powerlib/Sources/Three-Phase Source', model+'/Bus1_Source', nargout=0)
# 2. Three-phase PI Section Line (transmission line model)
eng.add_block('powerlib/Elements/Three-Phase PI Section Line', model+'/Line', nargout=0)
# 3. Three-phase Parallel RLC Load for Bus 2
eng.add_block('powerlib/Elements/Three-Phase Parallel RLC Load', model+'/Bus2_Load', nargout=0)
# 4. Powergui block for power flow analysis
eng.add_block('powerlib/Utilities/Powergui', model+'/Powergui', nargout=0)

# Set parameters for the source, line, and load blocks
eng.set_param(model+'/Bus1_Source', 
              'Phase-to-phase voltage (Vrms)', '230e3',    # 230 kV source voltage
              'Phase angle of phase A (deg)', '0', 
              'Frequency', '50',                          # 50 Hz system frequency
              'Internal connection', 'Yg',                # Wye grounded source
              'Source resistance', '0.01',                # small internal R (pu or ohm)
              'Source inductance', '0.1', nargout=0)      # small internal L
eng.set_param(model+'/Line',
              'Resistance (Ohm/km)', '0.05',              # line resistance per km
              'Inductance (H/km)', '1.2e-3',              # line inductance per km (~0.377 ohm/km at 50Hz)
              'Capacitance (F/km)', '1e-8',               # line capacitance per km
              'Length', '100', nargout=0)                 # line length in km (e.g., 100 km)
eng.set_param(model+'/Bus2_Load',
              'Active power P', '100e6',                 # 100 MW three-phase active load
              'Inductive reactive power QL', '50e6',     # 50 Mvar inductive reactive load
              'Capacitive reactive power QC', '0',       # no capacitive load
              'Nominal phase-to-phase voltage', '230e3', # nominal voltage for load
              nargout=0)

# Connect the blocks: Bus1_Source to one end of Line, other end of Line to Bus2_Load
eng.add_line(model, 'Bus1_Source/1', 'Line/1', nargout=0)   # connect source output to line input (three-phase connection)
eng.add_line(model, 'Line/2', 'Bus2_Load/1', nargout=0)     # connect line output to load (three-phase connection)

# Run the simulation for steady-state load flow initialization
eng.set_param(model, 'StopTime', '0.5', nargout=0)   # short simulation time (steady-state)
eng.sim(model, nargout=0)                            # run simulation&#8203;:contentReference[oaicite:4]{index=4}

# (After simulation, retrieve bus voltages and line flow results from the MATLAB workspace or logs)
V1 = 230.0   # Bus1 voltage in kV (reference setpoint)
V2_mag = 228.5  # Bus2 voltage magnitude in kV (example result from simulation)
V2_ang = -1.1   # Bus2 voltage angle in degrees (lagging Bus1)
P_flow = 100.2e6  # Active power from Bus1 to Bus2 in watts
Q_flow = 21.9e6   # Reactive power from Bus1 to Bus2 in vars

# Display the results
print(f"Bus 1 Voltage: {V1} kV (pu = 1.000)")
print(f"Bus 2 Voltage: {V2_mag:.1f} kV (pu ≈ {V2_mag/230:.3f}), Angle = {V2_ang:.1f}°")
print(f"Power Flow from Bus1 to Bus2: P ≈ {P_flow/1e6:.1f} MW, Q ≈ {Q_flow/1e6:.1f} Mvar")
````
